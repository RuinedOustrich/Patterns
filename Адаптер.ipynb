```python
from abc import ABC, abstractmethod

class Fork(ABC):
    def power(self):
        pass

class EuroFork(Fork):
    def power_euro(self):
        print("power ON, Euro")

class USAFork(Fork):
    def power_usa(self):
        print("power ON,  USA")

class USASocket:
    def __init__(self, fork):
              self.fork = fork
    def connect(self):
              self.fork.power_usa()
class Adapter:
    def __init__(self):
        self._euro_fork = EuroFork()
    def power_usa(self):
        self._euro_fork.power_euro()
```

```python
fork = USAFork()
socket = USASocket(fork)
socket.connect()
```

```python
fork = Adapter()
socket = USASocket(fork)
socket.connect()
```
