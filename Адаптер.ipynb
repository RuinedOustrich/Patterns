```python
from abc import ABC, abstractmethod

class Fork(ABC):
    def power(self):
        pass

class EuroFork(Fork):
    def power_euro(self):
        print("power ON, Euro")

class USAFork(Fork):
    def power_usa(self):
        print("power ON,  USA")

class USASocket:
    def __init__(self, fork):
              self.fork = fork
    def connect(self):
              self.fork.power_usa()
class Adapter:
    def __init__(self):
        self._euro_fork = EuroFork()
    def power_usa(self):
        self._euro_fork.power_euro()
```

```python
fork = USAFork()
socket = USASocket(fork)
socket.connect()

fork = Adapter()
socket = USASocket(fork)
socket.connect()
```

```python
class Lion:
    
    def roar(self):
        pass
    
class AfricanLion(Lion):
    
    def roar(self):
        return'Африканский лев рычит - РРРРРРРРРР'

class AsianLion(Lion):
    
    def roar(self):
        return'Азиатский лев рычит - РРРРРРРРРР'
        
class Hunter:
    
    def __init__(self, lion: Lion):
        self.lion = lion
    
    def hunt(self):
        print(f'Охотник слышит, как {self.lion.roar()} и прицеливается')
        
class Dog:
    
    def bark(self):
        return'Дикая собака лает - ГАВ ГАВ ГАВ'
        
class DogAdapter(Lion):
    
    def __init__(self, dog: Dog):
        self.dog = dog
        
    def roar(self):
        return self.dog.bark()
        
```

```python
lion = AfricanLion()
hunter = Hunter(lion)
hunter.hunt()

wilddog = Dog()
dog = DogAdapter(wilddog)
hunter = Hunter(dog)
hunter.hunt()
```
