```python
from abc import ABC, abstractmethod
```

```python
class Shirt(ABC):
    @abstractmethod
    def get_size(self):
        pass

class Pants(ABC):
    @abstractmethod
    def get_size(self):
        pass    
    
class Headdres(ABC):
    @abstractmethod
    def get_size(self):
        pass
    
class Tshirt(Shirt):
    def get_size(self):
        print('Размер футболки XL')
        
class Jeans(Pants):
    def get_size(self):
        print('Размер джинс XL')
        
class Cap(Headdres):
    def get_size(self):
        print('Размер кепки XL')
        
class ClassisShirt(Shirt):
    def get_size(self):
        print('Размер рубашки XL')
        
class ClassicPants(Pants):
    def get_size(self):
        print('Размер брюк XL')
        
class Hat(Headdres):
    def get_size(self):
        print('Размер шляпы XL')
        
class Store(ABC):
    @abstractmethod
    def get_shirt(self) -> Shirt:
        pass
    @abstractmethod
    def get_pants(self) -> Pants:
        pass
    @abstractmethod
    def get_headdress(self) -> Headdres:
        pass
    
class CasualStore(Store):
    
    def get_shirt(self) -> Tshirt:
        return Tshirt()
    def get_pants(self) -> Jeans:
        return Jeans()
    def get_headdress(self) -> Cap:
        return Cap()
    
class ClassicStore(Store):
    
    def get_shirt(self) -> ClassisShirt:
        return ClassisShirt()
    def get_pants(self) -> Jeans:
        return ClassicPants()
    def get_headdress(self) -> Cap:
        return Hat()

def show_clothes(store: Store):
    store.get_shirt().get_size()
    store.get_pants().get_size()
    store.get_headdress().get_size()
    
if __name__ == '__main__':
    show_clothes(CasualStore())
    show_clothes(ClassicStore())
```

```python
class Door(ABC):
    @abstractmethod
    def get_desc(self):
        pass

class Wooden_Door(Door):
    def get_desc(self):
        print("Деревянная дверь")

class Iron_Door(Door):
    def get_desc(self):
        print("Железная дверь")
        
class Expert(ABC):
    @abstractmethod
    def desc(self):
        pass

class WoodExpert(Expert):
    def desc(self):
        print("Плотник")
        
class IronExpert(Expert):
    def desc(self):
        print("Сварщик")
        
class Fabric(ABC):
    @abstractmethod
    def makedoor(self):
        pass
    @abstractmethod
    def sendexpert(self):
        pass
    
class WoodFabric(Fabric):
    def makedoor(self):
        return Wooden_Door()
    def sendexpert(self):
        return WoodExpert()

class IronFabric(Fabric):
    def makedoor(self):
        return Iron_Door()
    def sendexpert(self):
        return IronExpert()
    
def Zakaz(fabric: Fabric):
    fabric.makedoor().get_desc()
    fabric.sendexpert().desc()
    
def main():   
    Zakaz(WoodFabric())
    
if __name__ == "__main__":
    main()
```
