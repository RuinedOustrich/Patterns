```python
class Door:
    def open_(self):
        pass
    def close(self):
        pass

class LabDoor(Door):
    
    def open_(self):
        print('Открытие Двери Лаборатории...')
    def close(self):
        print('Закрытие Двери Лаборатории...')
        
class Security:
    
    def __init__(self, door: Door):
        self.door = door
        self.password = 12345'          
       
    def open_(self, password):
        if password == self.password:
            self.door.open_(self)
        else:
            print('пароль не верен')
            
    def close(self):
        self.door.close(self)
        
if __name__ == '__main__':
    door = Security(LabDoor)
    door.open_(2345)
    door.open_(12345)
    door.close()
```
