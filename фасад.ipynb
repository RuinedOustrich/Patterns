```python
class Computer:
    
    def getElectricShock(self):
        print("Подключается питание")
    
    def makeSound(self):
        print('Закгружается Жесткий Диск')
        
    def showLoading(self):
        print('Включается Экран')
        
    def bam(self):
        print('Готов к Использованию')
        
    def closeEverything(self):
        print('Выключаю Экран')
        
    def stop(self):
        print('Останавливаю работу')
    
    def off(self):
        print('Выключен')
        
class Facade:
    def __init__(self, computer: Computer):
        self.computer = computer
        
    def turnON(self):
        self.computer.getElectricShock()
        self.computer.makeSound()
        self.computer.showLoading()
        self.computer.bam()
        
    def turnOFF(self):
        self.computer.closeEverything()
        self.computer.stop()
        self.computer.off()
        
if __name__ == '__main__':
    
    computer = Facade(Computer())
    computer.turnON()
    computer.turnOFF()
```
