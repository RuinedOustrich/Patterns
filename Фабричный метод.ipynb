```python
from abc import ABC, abstractmethod

class User(ABC):
    pass

class MtsUser(User):
    
    def __init__(self, phone):
        print(f'Created MtsUser: {phone}')
        
class YandexUser(User):
    
    def __init__(self, email):
        print(f'Created YandexUser: {email}')
    
class Credentials(ABC):
    pass

class MtsCredentials(Credentials):
    
    def __init__(self, phone, password, email):
        self.phone = phone
        self.password = password
        self. email = email
    
    def email(self):
        return self.email
        
    def phone(self):
        return self.phone
    
    def password(self):
        return self.password
        
class YandexCredentials(Credentials):
    
    def __init__(self, email, password):
        self.email = email
        self.password = password

    def email(self):
        return self.email
    
    def password(self):
        return self.password
    
class Auth(ABC):
    
    def auth(self, credentials: Credentials) -> User:
        pass
    
class MtsAuth(Auth):
    
    def auth(self, credentials: MtsCredentials) -> MtsUser:
        print(f'authenticated by phone: {credentials.phone}')
        return MtsUser(credentials.phone)
    
class YandexAuth(Auth):
        
    def auth(self, credentials: YandexCredentials) -> YandexUser:
        print(f'authenticated by phone: {credentials.email}')
        return YandexUser(credentials.email)
        
phone = '+89966056562'
password = '123456'
email = 'ggg@mail.ru'
credentials = MtsCredentials(phone, password, email)
auth = MtsAuth()
user : MtsUser = auth.auth(credentials)
credentials.password
credentials.phone
credentials.email
```

```python
class ICar(ABC):
    def release(self):
        pass

class Car(ICar):
    def release(self):
        print("Выпуск легкого автомобиля")

class Truck(ICar):
    def release(self):
        print("Выпуск грузового автомобиля")
        
class Fabric(ABC):
    def create(self):
        pass
    
class Truc_Fabric(Fabric):
    def create(self):
        return Truck()
    
class Car_Fabric(Fabric):
    def create(self):
        return Car()
    
    
    
if __name__ == "__main__":
    creater = Truc_Fabric()
    truck = creater.create()
    
    creator = Car_Fabric()
    car = creator.create()

    car.release()
    truck.release()
```
