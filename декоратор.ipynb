```python
class Coffee:
    def getCost(self):
        pass
    def getDescription(self):
        pass
    
class SimpleCoffee(Coffee):
    def getCost(self):
        return 10
    def getDescription(self):
        return "Простой Коффе"
    
class Milk(Coffee):
    
    def __init__(self, coffee: Coffee):
        self.coffee = coffee
        
    def getCost(self):
        return self.coffee.getCost() + 2
    
    def getDescription(self):
        return f'{self.coffee.getDescription()} + молоко'
    
class Vanilla(Coffee):
    
    def __init__(self, coffee: Coffee):
        self.coffee = coffee
        
    def getCost(self):
        return self.coffee.getCost() + 3
    
    def getDescription(self):
        return f'{self.coffee.getDescription()} + ваниль'
```

```python
coffee = SimpleCoffee()
print(coffee.getCost())
print(coffee.getDescription())

coffee = Milk(coffee)
print(coffee.getCost())
print(coffee.getDescription())

coffee = Vanilla(coffee)
print(coffee.getCost())
print(coffee.getDescription())
```

```python
class Pizza:
    
    def get_cost(self):
        pass
    
    def ingridients(self):
        pass
    
class Margarita(Pizza):
        
    def get_cost(self):
        return 10
        
    def ingridients(self):
        return "Пицца Маргарита"
    
class Tomatoes:
    
    def __init__(self, pizza: Pizza):
        self.pizza = pizza
    
    def get_cost(self):
        return self.pizza.get_cost() + 2
    
    def ingridients(self):
        return f'{self.pizza.ingridients()} + Помидоры'
    
class Cheese:
    def __init__(self, pizza: Pizza):
        self.pizza = pizza
    
    def get_cost(self):
        return self.pizza.get_cost() + 5
    
    def ingridients(self):
        return f'{self.pizza.ingridients()} + Сыр'
```

```python
pizza = Margarita()
print(pizza.get_cost())
print(pizza.ingridients())

pizza = Tomatoes(pizza)
print(pizza.get_cost())
print(pizza.ingridients())

pizza = Cheese(pizza)
print(pizza.get_cost())
print(pizza.ingridients())
```
