```python
from abc import ABC, abstractmethod

class Builder(ABC):
    
    def build(self):
        self.test()
        self.lint()
        self.assemble()
        self.deploy()
        
    @abstractmethod
    def test(self):
        pass
    
    @abstractmethod
    def lint(self):
        pass
    
    @abstractmethod
    def assemble(self):
        pass
    
    @abstractmethod
    def deploy(self):
        pass
    
class AndroidBuilder(Builder):
    
    def test(self):
        print('Запуск Android тестов')
        
    def lint(self):
        print('Копирование Android кода')
        
    def assemble(self):
        print('Android сборка')
    
    def deploy(self):
        print('Развертывание сборки на сервере')
        
class IosBuilder(Builder):
    
    def test(self):
        print('Запуск iOS тестов')
        
    def lint(self):
        print('Копирование iOS кода')
        
    def assemble(self):
        print('iOS сборка')
    
    def deploy(self):
        print('Развертывание сборки на сервере')
        
if __name__ == '__main__':
    
   
    
    iosbuilder = IosBuilder()
    iosbuilder.build()
```

```python
androidbuilder = AndroidBuilder()
androidbuilder.build()
```
