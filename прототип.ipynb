```python
class Sheep:
        
    def set_name(self, name):
        self.name = name
    
    def set_category(self, category):
        self.category = category
        
    def get_name(self):
        print(self.name)
    
    def get_category(self):
        print(self.category)
```

```python
original = Sheep()
```

```python
original.set_name("Долли")
original.set_category("Горная овечка")
original.get_name()
original.get_category()
```

```python
sheep = original
sheep.set_name("Полли")
sheep.get_name()
sheep.get_category()
```
